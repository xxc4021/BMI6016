```python
import pandas as pd
```

```python
DW3 = pd.read_csv('DW3_set_exercise.csv',index_col= 0)
DW3['Patient ID'] = DW3['Patient ID'].astype(int)
DW3
```

```python
dia = ['E08','E09','E10','E11','E13']
dicode = DW3['Diagnosis Code'].str[:3]
diabetes = dicode.isin(dia)
db = DW3[diabetes]['Patient ID'].tolist()
print(set(db))
print(len(set(db)))
```

```python
covid = (DW3['Diagnosis Code'] == 'U07.1') | (DW3['Diagnosis Code'] == 'J12.82')
cov = DW3[covid]['Patient ID'].tolist()
print(set(cov))
print(len(set(cov)))
```

```python
inter = set(db).intersection(set(cov))
print(inter)
print(len(inter))
```

```python
union = set(db).union(set(cov))
print(union)
print(len(union))
```

```python
import matplotlib.pyplot as plt
from matplotlib_venn import venn2
# Use the venn2 function
venn2(subsets = (len(set(db))-len(set(inter)), len(set(cov))-len(set(inter)), len(set(inter))), set_labels = ('diabete', 'covid'))
plt.show()
```

```python
new_df = DW3[DW3['Patient ID'].isin(inter)]
new_df = new_df[new_df['Diagnosis Code'].str[:3].isin(dia) | new_df['Diagnosis Code'].isin(['U07.1', 'J12.82'])]
new_df['Date'] = pd.to_datetime(new_df['Date'])
new_df.reset_index(drop=True, inplace=True)
new_df['is_diabetes'] = new_df['Diagnosis Code'].str[:3].isin(dia)
new_df['is_covid'] = new_df['Diagnosis Code'].isin(['U07.1', 'J12.82'])
sorted_df = new_df.sort_values(by=['Patient ID', 'Date'])
first_diagnosis = sorted_df.groupby('Patient ID').first().reset_index()
covid_patients = first_diagnosis[first_diagnosis['is_covid']]['Patient ID'].tolist()
print(set(covid_patients))
print(len(set(covid_patients)))
covid_patients_df = new_df[new_df['Patient ID'].isin(covid_patients)]
covid_patients_df = covid_patients_df[covid_patients_df['is_diabetes']]
print(covid_patients_df['Diagnosis Code'].value_counts())
```
